## Normalisation do for all categorical columns

```python
import re
import string

def normalize_text(text):
    if pd.isnull(text):
        return ""
    # Lowercase
    text = text.lower()
    # Remove content inside parentheses
    text = re.sub(r'\([^)]*\)', '', text)
    # Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))
    # Remove extra whitespace
    text = re.sub(r'\s+', ' ', text).strip()
    # Remove url's whitespace
    text = re.sub(r'\s+', ' ', text).strip()
    return text

# Apply normalization to key fields for both datasets
ts_df['normalized_name'] = ts_df['name'].apply(normalize_text)
bd_df['normalized_product_name'] = bd_df['product_name'].apply(normalize_text)

# Display sample before and after normalization
sample_ts = ts_df[['name', 'normalized_name']].head(5)
sample_bd = bd_df[['product_name', 'normalized_product_name']].head(5)

sample_ts, sample_bd
```

```python
# Apply normalization to 'description' in ts_df and both 'description' and 'overview' in bd_df
ts_df['normalized_description'] = ts_df['description'].apply(normalize_text)
bd_df['normalized_description'] = bd_df['description'].apply(normalize_text)
bd_df['normalized_overview'] = bd_df['overview'].apply(normalize_text)

# Show sample cleaned descriptions
sample_ts_desc = ts_df[['description', 'normalized_description']].dropna().head(3)
sample_bd_desc = bd_df[['description', 'normalized_description']].dropna().head(3)
sample_bd_overview = bd_df[['overview', 'normalized_overview']].dropna().head(3)

sample_ts_desc, sample_bd_desc, sample_bd_overview
```

### Tokennisation (if further needed for NLP embeddings)

```python
import nltk
from nltk.tokenize import word_tokenize

# Ensure required NLTK resources are downloaded
nltk.download('punkt')

# Tokenization function
def tokenize_text(text):
    if not text:
        return []
    return word_tokenize(text)

# Apply tokenization to normalized name and description
ts_df['tokenized_name'] = ts_df['normalized_name'].apply(tokenize_text)
ts_df['tokenized_description'] = ts_df['normalized_description'].apply(tokenize_text)

bd_df['tokenized_name'] = bd_df['normalized_product_name'].apply(tokenize_text)
bd_df['tokenized_description'] = bd_df['normalized_description'].apply(tokenize_text)
bd_df['tokenized_overview'] = bd_df['normalized_overview'].apply(tokenize_text)

# Show samples
ts_df[['normalized_name', 'tokenized_name']].head(3), bd_df[['normalized_product_name', 'tokenized_name']].head(3)
```

### Stopword removal for description and overviews (Remove if it is not useful)

```python
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS

# Stopword removal function
def remove_stopwords(tokens):
    return [word for word in tokens if word not in ENGLISH_STOP_WORDS]

# Apply stopword removal to tokenized descriptions and overviews
ts_df['filtered_description_tokens'] = ts_df['tokenized_description'].apply(remove_stopwords)

bd_df['filtered_description_tokens'] = bd_df['tokenized_description'].apply(remove_stopwords)
bd_df['filtered_overview_tokens'] = bd_df['tokenized_overview'].apply(remove_stopwords)

# Show a few examples of before and after stopword removal
sample_ts_filtered = ts_df[['tokenized_description', 'filtered_description_tokens']].dropna().head(3)
sample_bd_filtered = bd_df[['tokenized_description', 'filtered_description_tokens']].dropna().head(3)
sample_bd_overview_filtered = bd_df[['tokenized_overview', 'filtered_overview_tokens']].dropna().head(3)

sample_ts_filtered, sample_bd_filtered, sample_bd_overview_filtered
```

## URL extraction

```python
from urllib.parse import urlparse

# Function to extract domain from URL
def extract_domain(url):
    if pd.isna(url) or not isinstance(url, str) or url.strip() == "":
        return ""
    parsed_url = urlparse(url)
    domain = parsed_url.netloc.lower().replace("www.", "")
    return domain

# Apply to both datasets
ts_df['domain'] = ts_df['url'].apply(extract_domain)
bd_df['domain'] = bd_df['seller_website'].apply(extract_domain)

# Show a few examples
ts_domains = ts_df[['url', 'domain']].dropna().head(5)
bd_domains = bd_df[['seller_website', 'domain']].dropna().head(5)

ts_domains, bd_domains
```
